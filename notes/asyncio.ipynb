```python
import asyncio
import time

async def task(name, delay):
    print(f"{name} started")
    """
    在这个task函数中,当执行到await行,会释放系统资源进而继续其他任务直到await返回
    """
    await asyncio.sleep(delay)
    print(f"{name} finished after {delay}s")

async def main():
    start = time.time()
    await asyncio.gather(
        task("A", 2),
        task("B", 2),
        task("C", 2),
    )
    print(f"Total time: {time.time() - start:.2f}s")

await main()  # 直接用 await，而不是 asyncio.run()
```
